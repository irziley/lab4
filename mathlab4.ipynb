```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
g = 7
w = 7 # w^2
```

```python
def f(t):
    return 7*np.sin(3*t)
```

```python
def dx(x, t):
    return np.array([x[1], -w*x[0]-g*x[1]-f(t)])
```

```python
t = np.linspace(0, 60, 600)
x0 = np.array([1, 2])

x = odeint(dx, x0, t)
```

```python
plt.plot(x[:, 0], x[:, 1])
plt.xlabel("x")
plt.ylabel("y")
plt.grid()
plt.show()
```
